```python
import pandas as pd
```

## Read data from csv files

```python
# import gene expression data from csv file
gene_expr = pd.read_csv("Example1_gene_expr.csv", index_col=0)
```

```python
# import raw count data from csv file
raw_count = pd.read_csv("Example1_raw_count.csv", index_col=0)
```

```python
# check imported gene expression levels
gene_expr
```

```python
# check imported raw counts
raw_count
```

## Inspect data

```python
# plot histrogram of gene expression (blue)
gene_expr.T.hist(figsize=(24,4), layout=(1,8), color="blue");
```

```python
# plot histrogram of raw counts (red)
raw_count.T.hist(figsize=(24,4), layout=(1,8), color="red");
```

```python
# plot boxplot for each gene (check the internal skewed distribution)

# gene expression data
gene_expr.T.boxplot(color="blue");
```

```python
# raw count
raw_count.T.boxplot(color="red");
```
